## Steps to Explore and Measure Data 

1. Examine the variables
2. Describe the dataset
3. Group the dataset in different ways (such as State)
4. Identify relationships (correlation) between variables
5. Graph some relationships between variables
6. Create one or more derived variables - suggested the difference between 1st income quartile life expectancy and 4th quartile life expectancy
6. Perform a linear regression predicting one variable on another

```python
def mulivariate_ols(formulatxt, df):
    mod = smf.ols(formula=formulatxt , data=df, missing='drop')
    res = mod.fit()
    print(res.summary())
    return res
```

```python
mulivariate_ols("le_raceadj_q4_q1_diff ~ inc_share_1perc +statename", health_inequality)
```

```python
mulivariate_ols("le_raceadj_q3_F ~ cur_smoke_q3 + gini99 + cs_frac_black + poor_share + inc_share_1perc + pop_density + cs_race_theil_2000 + frac_middleclass", health_inequality)
```
